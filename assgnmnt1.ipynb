```python
# ARITHMETIC OPERATIONS

n=int(input("enter a number\n"))
m=int(input("enter second number"))
while(1):
    print("1.sum\n2.subtract\n3.multiply\n4.divide\n5.exit)")
    ch=int(input("enter choice:"))
    if(ch==1):
        print("sum=",n+m)
    elif(ch==2):
        print("difference=",m-n)
    elif(ch==3):
        print("product=",n*m)
    elif(ch==4):
        print("quotient=",n/m)
    else:
        print("exiting...")
        break
    
```

```python
# STRING OPERATIONS

s1=str(input("enter the string1\n"))
s2=str(input("enter the string2\n"))
while(1):
    print("1.length of string2\n2.slicing of string\n3.to uppercase\n4.to lowercase\n5.replace\n6.concatenate\n7.exit)")
    ch=int(input("enter choice:"))
    if(ch==1):
        print("length=",len(s2))
    elif(ch==2):
        print(s1[2:])
    elif(ch==3):
        print("uppercase=",s1.upper())
    elif(ch==4):
        print("lowercase=",s2.lower())
    elif(ch==5):
        print("after replacing=",s2.replace("o","r")) 
    elif(ch==6):
        s=s1+s2
        print("after concatinating=",s)  
    else:
        print("exiting...")
        break
    


```

```python
# LIST OPERATIONS

a=str(input("enter the list:"))
b=a.split(",")
print(b)
print("first item=",b[0])
print("last item=",b[-1])
print("second,third and fourth item=",b[2:5])
print("change item=",b[2]=="apple")
print("add an item",b.append("strawberry"))
print(b)
print("remove an item ",b.remove("kiwi"))
print(b)
print("sorted list is",b.sort())
print(b)

```

```python
# DICTIONARY


dict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(dict["brand"])
print(len(dict))
x = dict["model"]
print(x)
dict["year"] = 2018
print(dict)
dict["color"] = "red"
print(dict)
dict.update({"style": "vintage"})
print(dict)
dict.pop("model")
print(dict) 
for x in dict:
    print(dict[x]) 
```
